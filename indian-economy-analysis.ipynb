### GDP of India has seen a steady growth since 2005
Since 2005, India's Gross Domestic Product (GDP) has exhibited a combination of robust growth and occasional economic challenges. The country experienced a period of steady expansion during the mid-2000s and early 2010s, positioning itself as one of the world's fastest-growing major economies.

# Ratio of Export vs Import

India has generally maintained a trade deficit, which means its imports have exceeded its exports. The ratio of exports to imports, commonly known as the trade balance, fluctuated over the years due to various economic factors and global trends. Until 1991, India followed a relatively protectionist trade policy, leading to limited export opportunities.

However, post-1991 economic reforms, which liberalized the Indian economy, set the stage for increased exports. The country experienced steady growth in exports over the following decades. Despite this growth, imports also surged, often driven by the rising domestic demand for consumer goods, raw materials, and capital equipment.

```python
india_imp_exp_ratio.plot(title='Export vs Imports',figsize = (14,5),x='year',y='Value',color='#8d0d87d0',lw=4)
```

In recent years, India has been trying to address its trade imbalance through various initiatives like "Make in India" and increasing focus on export-oriented sectors. By promoting manufacturing, services, and technology-driven industries, the country aims to boost exports and reduce its reliance on imports.

# Comparing INR vs USD, A direct comparison of strength over one another

Since 1970, the trend of the Indian Rupee (INR) against the United States Dollar (USD) has been dynamic, reflecting the ever-changing economic landscape and global market conditions. The INR was initially pegged to a fixed exchange rate against the USD, but India's economic reforms in 1991 led to a transition to a more flexible exchange rate system.

Throughout the years, the INR has experienced periods of appreciation and depreciation against the USD, driven by various factors like international economic trends, trade dynamics, capital flows, and domestic economic performance.
