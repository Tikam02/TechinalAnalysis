```python
# Define the path to your input and output text files
input_file_path = 'niftysmallcap500.txt'
output_file_path = 'formatted_niftysmall500.txt'


# Read stock symbols from the input text file into a single line
with open(input_file_path, 'r') as input_file:
    stock_symbols_line = input_file.read()

# Split the line into a list of stock symbols
stock_symbols_list = stock_symbols_line.split()

# Print the resulting list vertically
print("Stock Symbols List:")
for symbol in stock_symbols_list:
    print(symbol)

# Write the list to the output text file
with open(output_file_path, 'w') as output_file:
    for symbol in stock_symbols_list:
        output_file.write(symbol + '\n')

print(f"Stock symbols list has been written to {output_file_path}")
```

```python
# Define the path to your input text file
input_file_path = './formatted_niftysmall500.txt'
output_file_path = 'final_small500.txt'

# Read stock symbols from the input text file into a string
with open(input_file_path, 'r') as input_file:
    input_symbols = input_file.read()

# Remove single quotes and commas
cleaned_symbols = input_symbols.replace("'", "").replace(",", "")

# Split the cleaned string into a list of stock symbols
stock_symbols_list = cleaned_symbols.split()

# Print the resulting list vertically
print("Stock Symbols List:")
for symbol in stock_symbols_list:
    print(symbol)

# Write the list to the output text file
with open(output_file_path, 'w') as output_file:
    for symbol in stock_symbols_list:
        output_file.write(symbol + '\n')

print(f"Stock symbols list has been written to {output_file_path}")
```
