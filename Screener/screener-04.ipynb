```python
import yfinance as yf
import pandas as pd

def get_stock_data(symbol):
    ticker = yf.Ticker(symbol)
    data = ticker.history(period="1y")
    return data

def sma_crossover(data, short_window, long_window):
    data['SMA_Short'] = data['Close'].rolling(window=short_window).mean()
    data['SMA_Long'] = data['Close'].rolling(window=long_window).mean()
    return data['SMA_Short'].iloc[-1] > data['SMA_Long'].iloc[-1]

def filter_stocks(file_path):
    selected_stocks = []

    with open(file_path, 'r') as file:
        stock_symbols = [symbol.strip() for symbol in file.readlines()]

    for symbol in stock_symbols:
        data = get_stock_data(symbol)

        # Condition 1: 44 SMA crosses 200 SMA
        condition_sma_44_200 = sma_crossover(data, 44, 200)

        # Condition 2: 25 SMA crosses 44 SMA
        condition_sma_25_44 = sma_crossover(data, 25, 44)

        # Check if any of the conditions are met
        if condition_sma_44_200 or condition_sma_25_44:
            selected_stocks.append(symbol)

    return selected_stocks

# Specify the path to the file
file_path = './formatted_stock_symbols.txt'

# Filter stocks based on conditions
selected_stocks = filter_stocks(file_path)

# Display selected stocks
print("Selected Stocks:")
for stock in selected_stocks:
    print(stock)
```
