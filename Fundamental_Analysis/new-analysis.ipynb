```python
import requests
from bs4 import BeautifulSoup

url = "https://pulse.zerodha.com/"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content
    html_content = response.content
    soup = BeautifulSoup(html_content, 'html.parser')

    # Save the parsed HTML content to a text file
    with open('parsed_html_content.txt', 'w', encoding='utf-8') as file:
        file.write(soup.prettify())

    # Extract dates from the webpage
    dates = [span['title'] for span in soup.select('.news-ticker .date')]

    # Display the extracted dates
    print("Extracted Dates:")
    for date in dates:
        print(date)

else:
    print(f"Failed to retrieve the content. Status Code: {response.status_code}")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://pulse.zerodha.com/"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content
    html_content = response.content
    soup = BeautifulSoup(html_content, 'html.parser')

    # Extracting data from the HTML
    data_list = []
    for news_item in soup.find_all('li', class_='box item'):
        title = news_item.find('h2', class_='title').text.strip()
        description = news_item.find('div', class_='desc').text.strip()
        date = news_item.find('span', class_='date')['title']
        
        # Extracting source link
        source_link = news_item.find('a', class_='tw')['href'] if news_item.find('a', class_='tw') else None

        data_list.append({'Title': title, 'Description': description, 'Date': date, 'Source_Link': source_link})

    # Creating a DataFrame
    df = pd.DataFrame(data_list)

    # Saving the DataFrame to a CSV file
    df.to_csv('extracted_data.csv', index=False)

    # Displaying the DataFrame
    print(df)
else:
    print(f"Failed to retrieve the content. Status Code: {response.status_code}")
```

```python
import pandas as pd
from bs4 import BeautifulSoup

# Assuming you have the HTML content stored in a file named 'parsed_html_content.txt'
with open('extracted_data.csv', 'r', encoding='utf-8') as file:
    html_content = file.read()

soup = BeautifulSoup(html_content, 'html.parser')

# Extracting data from the HTML
data_list = []
for news_item in soup.find_all('li', class_='box item'):
    title = news_item.find('h2', class_='title').text.strip()
    description = news_item.find('div', class_='desc').text.strip()
    date = news_item.find('span', class_='date')['title']

    # Check if there's a link with class 'fb'
    link_element = news_item.find('a', class_='fb')
    if link_element:
        link = link_element['href']
    else:
        link = None

    # Check if there's a 'Source_Link' column in the DataFrame
    source_link_element = news_item.find('td', class_='Source_Link')
    if source_link_element:
        source_link = source_link_element.text.strip()
    else:
        source_link = None

    data_list.append({'Title': title, 'Description': description, 'Date': date, 'Link': link, 'Source_Link': source_link})

# Creating a DataFrame
df = pd.DataFrame(data_list)

# Print the column names to identify the correct case
print(df.columns)

# Searching for a specific company (e.g., LIC in this case)
company_name = 'LIC'

# Adjust the column names based on the actual case in your DataFrame
search_results = df[df['Title'].str.contains(company_name, case=False) | df['Description'].str.contains(company_name, case=False)]

# Extracting the 'Source_Link' column associated with the searched company
source_links = search_results['Source_Link']

# Displaying the 'Source_Link' column
print("Source Links for", company_name)
print(source_links)
```
