```python
import pandas as pd
import json

def json_to_csv(json_file, csv_file):
    try:
        # Load JSON file
        with open(json_file, 'r') as f:
            data = json.load(f)
        
        # Convert JSON to DataFrame
        df = pd.json_normalize(data)
        
        # Save DataFrame to CSV
        df.to_csv(csv_file, index=False)
        return True
    except Exception as e:
        print(f"Error converting JSON to CSV: {e}")
        return False

# Example usage
json_file = "most_active_monthly.json"
csv_file = "most_active_monthly.csv"
success = json_to_csv(json_file, csv_file)

if success:
    print(f"JSON file '{json_file}' converted to CSV file '{csv_file}' successfully.")
else:
    print("Conversion failed.")
```

```python
import json
import csv

# Step 1: Read the JSON file
def read_json_file(file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)
    return data['data'] # Access the list under the 'data' key

# Step 2: Extract the keys
def extract_keys(data):
    keys = data[0].keys()
    return keys

# Step 3: Write the CSV file
def write_csv_file(file_path, data, keys):
    with open(file_path, 'w', newline='') as file:
        writer = csv.DictWriter(file, fieldnames=keys)
        writer.writeheader()
        for row in data:
            writer.writerow(row)

# Main function to convert JSON to CSV
def json_to_csv(json_file_path, csv_file_path):
    data = read_json_file(json_file_path)
    keys = extract_keys(data)
    write_csv_file(csv_file_path, data, keys)

# Example usage
json_file_path = './most_active_monthly.json'
csv_file_path = './output.csv'
json_to_csv(json_file_path, csv_file_path)
```

```python
import requests
import json
import csv

# Function to download JSON data from a URL
def download_json_data(url, headers=None):
    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status() # Raise an exception for bad status codes
        return response.json()
    except requests.exceptions.RequestException as e:
        print("Error fetching data:", e)
        return None

# Function to save JSON data to a file
def save_json_to_file(data, file_path):
    with open(file_path, 'w') as file:
        json.dump(data, file)
    print(f"Data saved to '{file_path}'")

# Function to convert JSON to CSV
def json_to_csv(json_file_path, csv_file_path):
    with open(json_file_path, 'r') as file:
        data = json.load(file)
    keys = data[0].keys()
    with open(csv_file_path, 'w', newline='') as file:
        writer = csv.DictWriter(file, fieldnames=keys)
        writer.writeheader()
        for row in data:
            writer.writerow(row)
    print(f"Data converted to CSV and saved to '{csv_file_path}'")

if __name__ == "__main__":
    url = " https://www.nseindia.com/api/block-deal"
    headers = {
        "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36",
        "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8",
        "Accept-Encoding": "gzip, deflate, br",
        "Accept-Language": "en-GB,en;q=0.5",
        "Cookie": "Your Cookie Value Here",
        # Add any other necessary headers here
    }
    json_data = download_json_data(url, headers=headers)
    
    if json_data:
        # Save JSON data to a file
        save_json_to_file(json_data, './most_active_monthly.json')
        
        # Convert JSON to CSV
        json_to_csv('./most_active_monthly.json', './output.csv')
```
