Mean:

Explanation: In statistics, the mean is the average of a set of values. It is calculated by adding up all the values and then dividing by the number of values.

```python
import yfinance as yf

# Download historical data for ITC.NS
itc_data = yf.download('DEEPAKFERT.NS', start='2023-01-01', end='2024-01-01')

# Calculate mean (average) of the closing prices
mean_value = itc_data['Close'].mean()

print("Mean:", mean_value)
```

### Plot mean

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Download historical data for ITC.NS
itc_data = yf.download('DEEPAKFERT.NS', start='2023-06-01', end='2024-01-31')

# Calculate mean (average) of the closing prices
mean_value = itc_data['Close'].mean()

# Plot the closing prices and the mean
plt.figure(figsize=(10, 6))
plt.plot(itc_data['Close'], label='DEEPAKFERT.NS Close Price')
plt.axhline(y=mean_value, color='r', linestyle='--', label='Mean')
plt.legend()
plt.title('DEEPAKFERT.NS Stock Prices with Mean')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.show()
```

### Mean Reversion:

Explanation: Mean reversion is a financial theory that suggests that asset prices and returns eventually move back toward the mean or average over time. It implies that if the price of an asset deviates significantly from its historical average, it is likely to revert to that average.

```python
# Calculate mean of the closing prices
mean_value = itc_data['Close'].mean()

# Check if the current closing price is above or below the mean
current_price = itc_data['Close'].iloc[-1]

if current_price > mean_value:
    print("Price is above the mean - consider selling")
elif current_price < mean_value:
    print("Price is below the mean - consider buying")
else:
    print("Price is at the mean - hold")
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Download historical data for ITC.NS
itc_data = yf.download('ITC.NS', start='2023-01-01', end='2024-01-01')

# Calculate a simple moving average (e.g., 20 days)
itc_data['SMA_20'] = itc_data['Close'].rolling(window=20).mean()

# Plot the closing prices and the moving average
plt.figure(figsize=(10, 6))
plt.plot(itc_data['Close'], label='ITC.NS Close Price')
plt.plot(itc_data['SMA_20'], label='20-Day SMA', color='orange')

plt.legend()
plt.title('ITC.NS Stock Prices with Mean Line')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.show()
```

### Standard Deviation:

Explanation: Standard deviation measures the amount of variation or dispersion of a set of values. In finance, it is often used as a measure of volatility. A higher standard deviation indicates higher volatility.

```python
# Calculate standard deviation of the closing prices
std_deviation = itc_data['Close'].std()

print("Standard Deviation:", std_deviation)
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Download historical data for ITC.NS
itc_data = yf.download('ITC.NS', start='2023-01-01', end='2024-01-01')

# Calculate standard deviation of the closing prices
std_deviation = itc_data['Close'].rolling(window=20).std()

# Plot the closing prices, the mean line, and the standard deviation bands
plt.figure(figsize=(10, 6))
plt.plot(itc_data['Close'], label='ITC.NS Close Price')
plt.plot(itc_data['Close'].rolling(window=20).mean(), label='20-Day SMA', color='orange')
plt.fill_between(itc_data.index, itc_data['Close'] - std_deviation, itc_data['Close'] + std_deviation, color='lightgray', label='Standard Deviation Band')

plt.legend()
plt.title('ITC.NS Stock Prices with Standard Deviation Bands')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.show()
```

### Moving Average:

Explanation: A moving average is a statistical calculation used to analyze data over a certain period. In finance, the moving average is often used to smooth out price data and identify trends by creating a single flowing line.

```python
# Calculate a simple moving average (SMA) with a window of 20 days
itc_data['SMA_20'] = itc_data['Close'].rolling(window=20).mean()

# Plot the closing prices and the moving average
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(itc_data['Close'], label='ITC.NS Close Price')
plt.plot(itc_data['SMA_20'], label='20-Day SMA', color='orange')
plt.legend()
plt.show()
```
